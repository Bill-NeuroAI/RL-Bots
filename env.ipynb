# Environment 1
We will first implement 10 slot machines.


Environment owns 3 things. True rewards, probabilities, and step function

```python
mu = jnp.array([-1.0, -0.5, 0.0, 0.3, 0.6, 0.9, 1.2, 1.0, 0.4, -0.8])
#state is (t,N,Q)
def step_slot(state, t_r):
    if not (jnp.any(state[1] == 0)):
        return 
    else:
        N = state[1]
        mask = (N ==0)
        f_index = mask.argmax()
        N = N.at[f_index].set(N[f_index]+1)

        Q = state[2]
        Q = Q.at[f_index].set(Q[f_index] + t_r[f_index])

        t = state[0] + 1
        state = (t, N, Q)
        return state


```
